# Standard Normal Distribution in Python

## Importing Relevant Libraries

```python
import numpy as np
import pandas as pd
```

```python
#need scipy and the norm libraries in order to calculate the probability distribution function (pdf), cumulative distribution function (cdf) and ppf
import scipy as sy
from scipy.stats import norm
```

## Probability Distribution Function

```python
#Calculate probability given z-score
z = norm.pdf(1.96)
print(z)
```

## Cumulative Distribution Function

```python
#Calculate CDF given z-score
y = norm.cdf(1.96)
print(y)
```

```python
#Note that the inverse is equal
y1 = norm.cdf(-1.96)
print(y1)
print(1-y1)
```

## Point Percentile Function

```python
#Calculate z-score given probability
x = norm.ppf(0.975)
print(x)
```

## Plotting a Normal Distribution

```python
#To make plots, recommend to import these two libraries, seaborn & matplotlib as they provide extra functionality for visualization
import seaborn as sns
import matplotlib.pyplot as plt

#arange makes "a range of data" with the following inputs: (min value, max value, step)  
x_B = np.arange(70,100,.01)

#takes the data points from x_B and fits them to a pdf(dataframe,loc=mean,scale=stdev)
normal_B=norm.pdf(x_B,loc=85,scale=4)

#makes a lineplot
ax=sns.lineplot(x=x_B,y=normal_B)

#adding shading to demonstrate probabilities
ax=plt.fill_between(x_B,normal_B, where=(x_B <77),color='k',alpha=.99)
ax=plt.fill_between(x_B,normal_B, where=(x_B <93), color='k', alpha=.2)

#Labels, xlabel for x axis, ylabel would be for y axis
plt.xlabel("X=batch yields (g/L)")
plt.title("Normal Distribution with Mean=85, SD=4")

#Colours for your reference:     
    #'b'- blue.
    #'c' - cyan.
    #'g' - green.
    #'k' - black.
    #'m' - magenta.
    #'r' - red.
    #'w' - white.
    #'y' - yellow.
```
